### Funções de Bessel do primeiro tipo

As funções de Bessel do primeiro tipo, na sua forma integral, são dadas por

$$
J_m(x) = \frac 1 \pi \int_0^\pi \cos \left( m\theta - x \sin \theta \right)\ \mathrm{d}\theta,
$$

onde $m$ é um inteiro não negativo, e $x \geq 0$.

#### Item 1

Escreva uma função que calcule o valor de $J_m(x)$ usando o método de quadratura gaussiana (use $N=20$).

```python
import numpy as np
from scipy.special import roots_legendre

N = 20 #Valor para uma boa aproximacao como visto nos ultimos exercicios
limInf = 0.0 #essa integral sempre de 0 ate outro valor
limSup = np.pi

def integrando(teta, x, m):
    return np.cos(m * teta - x * np.sin(teta))

def J(m, x): #O que varia eh apenas o valor de x no integrando e o m
    I = 0.0
    u_k, w_linha_k = roots_legendre(N)

    for i in range(N):
        w_k = (limSup - limInf) * w_linha_k[i] / 2
        x_k = ((limSup - limInf) * u_k[i] / 2) + ((limSup + limInf) / 2)
        I = I + w_k * integrando(x_k, x, m)
    return I / np.pi
```

#### Item 2

Use sua função para fazer um gráfico comparando $J_0$, $J_1$ e $J_2$ (no mesmo gráfico), com $0 \leq x \leq 20$.

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.rcParams['font.family'] = 'serif'
plt.rcParams['mathtext.fontset'] = 'stix'

lista_N = []
lista_I_0 = [] #onde 0, 1, e 2 sao os valores de m de cada funcao
lista_I_1 = []
lista_I_2 = []

x = 0.0
while (x <= N):
    lista_N.append(x)
    lista_I_0.append(J(0, x))
    lista_I_1.append(J(1, x))
    lista_I_2.append(J(2, x))
    x += 0.1

fig, ax = plt.subplots()
ax.axhline(0.0, ls=':', color='k')
ax.plot(lista_N, lista_I_0, label="J0")
ax.plot(lista_N, lista_I_1, label="J1")
ax.plot(lista_N, lista_I_2, label="J2")
ax.set_title("Funções de Bessel do primeiro tipo")
ax.legend()
plt.show()
```
